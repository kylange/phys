```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = ['5 kg','10 kg','15 kg']
dom = [0.044,0.081,0.166]
```

```python
X_axis = np.arange(len(X))
  
plt.bar(X_axis, dom, label = 'Condition of Apnea')
  
plt.xticks(X_axis, X)
plt.xlabel("Force (kg)")
plt.ylabel("EMG reading (mv/sec)")
plt.title("Effect of Force on EMG Reading")
plt.legend()
plt.show()
```
